```python
from google.colab import drive
drive.mount('/content/drive')

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from skimage.morphology import skeletonize
from skimage.util import invert

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
excelsa_dir = '/content/drive/MyDrive/LibericaxExcelsa/Excelsa'
liberica_dir = '/content/drive/MyDrive/LibericaxExcelsa/Liberica'
image_size = (224, 224)
label_map = {'Excelsa': 0, 'Liberica': 1}
```

```python
def calculate_eccentricity(contour):
    if len(contour) < 5:
        return 0
    ellipse = cv2.fitEllipse(contour)
    major = max(ellipse[1])
    minor = min(ellipse[1])
    if major == 0:
        return 0
    return np.sqrt(1 - (minor / major) ** 2)

def preprocess_and_extract(image_path):
    image = cv2.imread(image_path)
    if image is None:
        return None, None

    image = cv2.resize(image, image_size)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    _, thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Invert background if needed
    if np.mean(thresh) > 127:
        thresh = cv2.bitwise_not(thresh)

    # Morphological closing to remove noise and fill gaps
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
    closed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    # Find contours
    contours, _ = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return None, None

    c = max(contours, key=cv2.contourArea)

    # Extract only the bean (background removed)
    mask = np.zeros_like(gray)
    cv2.drawContours(mask, [c], -1, 255, -1)
    bean_only = cv2.bitwise_and(image, image, mask=mask)

    # Bean shape and size
    x, y, w, h = cv2.boundingRect(c)
    area = cv2.contourArea(c)
    aspect_ratio = float(w) / h
    perimeter = cv2.arcLength(c, True)
    circularity = 4 * np.pi * (area / (perimeter**2)) if perimeter != 0 else 0
    eccentricity = calculate_eccentricity(c)

    # Hu Moments
    hu = cv2.HuMoments(cv2.moments(c)).flatten()
    hu_log = -np.sign(hu) * np.log10(np.abs(hu) + 1e-10)

    # Tip sharpness via hull angles
    hull = cv2.convexHull(c, returnPoints=True)
    angles = []
    for i in range(len(hull)):
        p1 = hull[i - 1][0]
        p2 = hull[i][0]
        p3 = hull[(i + 1) % len(hull)][0]
        v1 = p1 - p2
        v2 = p3 - p2
        angle = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2) + 1e-10))
        angles.append(np.degrees(angle))
    tip_sharpness = np.mean(sorted(angles)[:2])

    # Skeleton and curve (centerline)
    binarized = (closed > 0).astype(np.uint8)
    skeleton = skeletonize(invert(binarized))
    coords = np.argwhere(skeleton)

    if coords.shape[0] < 2:
        coeffs = [0, 0, 0]
        curve_strength = 0
    else:
        curve_x = coords[:,1]
        curve_y = coords[:,0]
        coeffs = np.polyfit(curve_x, curve_y, deg=2)
        curve_strength = abs(coeffs[0])  # Key S-curve feature

    # Final feature vector
    features = [
        curve_strength,               # 🧠 Key S-curve feature
        area, w, h, aspect_ratio,     # Size + Shape
        circularity, eccentricity, tip_sharpness
    ] + coeffs.tolist() + hu_log.tolist()

    return bean_only / 255.0, features
```

```python
X_shape, y = [], []

def process_folder(folder_path, label_name):
    label = label_map[label_name]
    for file in tqdm(os.listdir(folder_path), desc=f"Processing {label_name}"):
        if not file.lower().endswith(('.jpg', '.png')):
            continue
        path = os.path.join(folder_path, file)
        _, features = preprocess_and_extract(path)
        if features is not None:
            X_shape.append(features)
            y.append(label)
        else:
            print(f"⚠️ Skipped {file}: no features extracted")

process_folder(liberica_dir, 'Liberica')
process_folder(excelsa_dir, 'Excelsa')
```

```python
if len(X_shape) == 0:
    raise ValueError("❌ No features extracted.")

X_shape = StandardScaler().fit_transform(np.array(X_shape))
y = np.array(y)

X_shape_train, X_shape_val, y_train, y_val = train_test_split(
    X_shape, y, test_size=0.7, stratify=y, random_state=42
)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_shape_train, y_train)
```

```python
y_val_pred = rf_model.predict(X_shape_val)
print(classification_report(y_val, y_val_pred, target_names=['Excelsa', 'Liberica']))

sns.heatmap(confusion_matrix(y_val, y_val_pred), annot=True, fmt='d',
            xticklabels=['Excelsa', 'Liberica'], yticklabels=['Excelsa', 'Liberica'])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix - Random Forest")
plt.show()
```

```python
from google.colab import files
from IPython.display import Image as IPImage, display

# Upload image
uploaded = files.upload()
img_path = list(uploaded.keys())[0]
display(IPImage(filename=img_path))

# S-Curve + Random Forest prediction function
def predict_rf(img_path, threshold=0.8, curve_cutoff=0.4999999999997):
    _, features = preprocess_and_extract(img_path)
    if features is None:
        print("❌ Could not extract features.")
        return

    shape_input = np.expand_dims(features, axis=0)
    proba = rf_model.predict_proba(shape_input)[0]
    classes = ['Excelsa', 'Liberica']

    # Extract S-curve strength (first feature in our design)
    curve_strength = features[0]

    print("\n🔍 Prediction Confidence (from Random Forest):")
    for i, score in enumerate(proba):
        print(f"  {classes[i]}: {score * 100:.2f}%")

    print(f"\n📐 S-Curve Strength: {curve_strength:.6f} (Cutoff: {curve_cutoff})")

    # Rule-based override using S-curve
    if curve_strength >= curve_cutoff:
        print("🎯 Final Decision: Based on strong S-curve → Predicted as ➤ Liberica")
    elif curve_strength < curve_cutoff:
        print("🎯 Final Decision: Based on straight/slight curve → Predicted as ➤ Excelsa")
    elif np.max(proba) >= threshold:
        print(f"✅ Final Decision: Model predicts ➤ {classes[np.argmax(proba)]} (≥ {threshold*100:.0f}%)")
    else:
        print("❓ Final Decision: Other (Uncertain, low confidence)")
```
